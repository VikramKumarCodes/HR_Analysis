# HR ANALYSIS: 

## In this module, we describe how to analyze a HR Spreadsheet for those who are applying for post of Data Analyst and Data Scientist in a Company

```python
# Importing the required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Importing csv file containing form entries
hr_data= pd.read_csv('Spreadsheet.csv')
```

```python
#Displaying the Top 5 rows in Data Set
hr_data.head()
```

# DATA PREPROCESSING:

```python
#Check if any Column's has NA or Missing Values
hr_data.isnull().any()
```

```python
#Shape of the Data
hr_data.shape
```

```python
#Data Types of Columns
hr_data.dtypes
```

```python
hr_data.columns = ['Timestamp','Name','Gender','Age','Experience','Job_Role','Python_Skill','R_Skill','SQL_Skill','Advanced_Excel_Skill','Data_Visualization_Skill','Machine_Learning_Skill','SAS_Skill','Apache_Spark_Skill','AWS_Skill','Deep_Learning_Skill']
```

```python
hr_data.set_index('Timestamp',inplace=True)
```

```python
hr_data
```

# EXPLORATORY DATA ANALYSIS:

```python
#Percentage Distribution
hr_data.Job_Role.value_counts(normalize=True)
```

```python
plt.figure(figsize=(8,6))
Job=hr_data.Job_Role.value_counts()
sns.barplot(x=Job.index ,y=Job.values)
plt.title('Distribution of Employee ')
plt.xlabel('Employee Type', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

```python
hr_data.groupby(["Job_Role","Gender"]).size().unstack().plot(kind = "bar", stacked = True,figsize=(8,8))
plt.title("STACKED BAR CHART FOR GENDER")
plt.xlabel("JOB ROLE")
```

```python
df1=hr_data[hr_data["Job_Role"]=='Data Scientist']
```

```python
df2=hr_data[hr_data["Job_Role"]=='Data Analyst']
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(df1['Age'])
plt.title("DENSITY GRAPH FOR AGE WISE DISTRIBUTION OF DATA SCIENTIST")
plt.ylabel("DENSITY")
plt.xlabel("AGE")
df1['Age'].mean()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(df2['Age'])
plt.title("DENSITY GRAPH FOR AGE WISE DISTRIBUTION OF DATA ANALYST")
plt.ylabel("DENSITY")
plt.xlabel("AGE")
df2['Age'].mean()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(df1['Experience'])
plt.title("DENSITY GRAPH FOR EXPERIENCE WISE DISTRIBUTION OF DATA SCIENTIST")
plt.ylabel("DENSITY")
plt.xlabel("EXPERIENCE")
df1['Experience'].mean()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(df2['Experience'])
plt.title("DENSITY GRAPH FOR EXPERIENCE WISE DISTRIBUTION OF DATA ANALYST")
plt.ylabel("DENSITY")
plt.xlabel("EXPERIENCE")
df2['Experience'].mean()


```
